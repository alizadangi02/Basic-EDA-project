**Exploratory Data Analysis (EDA)**

Dataset: U.S. Top 50 Universities 2004_2026

Link: https://www.kaggle.com/datasets/nalisha/u-s-top-50-universities-2004-2026

```python
#importing libraries
import pandas as pd #for data handling
import numpy as np #for numerical operations
#for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv') #load dataset
```

```python
#checks first 5 rows
df.head()
```

```python
#shows rows and columns
df.shape
```

```python
#only columns
df.columns
```

```python
#data types and null values
df.info()
```

```python
#statistical summary such as mean, std, min, max
df.describe()
```

```python
#check missing values
df.isnull().sum()
```

```python
#drop missing values
df = df.dropna()
```

```python
#check duplicate data
df.duplicated().sum()
```

```python
#remove duplicates if found to maintain DATA INTEGRITY
df.drop_duplicates(inplace=True)
```

**Visualize Data Distribution**

```python
#Histograms
df.hist(figsize=(12,8))
plt.show()

#shows distribution shape, skewness and spread of data
```

```python
#Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()

#shows outliers, median and range
```

```python
#Correlation Analysis
plt.figure(figsize=(8,6))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
#shows:
#strong positive correlation --> close to +1
#strong negative correlation --> close to -1
#Weak correlation --> close to 0
```

```python
#Pairplot
sns.pairplot(df)
plt.show()
#shows patterns, linear relationships and clusters.
```
